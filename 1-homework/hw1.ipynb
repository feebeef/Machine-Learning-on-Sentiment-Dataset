```python
file  = open("starwars.txt", "r") 
x = file.read()

#https://www.debuggex.com/cheatsheet/regex/python
#https://pythex.org/
import re
words = re.findall("(\w+-*\w*'*\w*)",x)
print('total: ' + str(len(words)))
```

```python
print(words)
```

```python
frequency = {}

for word in words:
    count = frequency.get(word.lower(), 0)
    frequency[word.lower()] = count + 1
```

```python
frequency_list = frequency.keys()
```

```python
for bananas in frequency_list:
    print (bananas, frequency[bananas])
```

```python
luke = re.findall("(LUKE\n[\s\S]+?\n\n)", x)
luke = re.findall("(LUKE\n[\w\W]+?\n\n)", x)

frequency = {}
for dialog in luke:
    words = re.findall("(\w+-*\w*'*\w*)",dialog)
    
    for word in words:
        count = frequency.get(word.lower(), 0)
        frequency[word.lower()] = count + 1
        count = frequency.get(word.lower(), 0)
        frequency[word.lower()] = count + 1


frequency_list = frequency.keys()
print("number of luke vocab: " + str(len(frequency_list))
      
frequency_list = frequency.keys()
print("number of luke vocab: " + str(len(frequency_list))
      

for bananas in frequency_list:
    print (bananas, frequency[bananas])
```

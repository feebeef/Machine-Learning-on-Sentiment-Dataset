```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

train_csv = pd.read_csv('train.csv', error_bad_lines=False).fillna(' ')
test_csv = pd.read_csv('test.csv', error_bad_lines=False).fillna(' ')
display(train_csv.head())

# aggregation of toxic labels
train = train_csv
train['general_toxic'] = ((train['toxic'] + train['severe_toxic'] + train['obscene'] + train['threat'] + train['insult'] + train['identity_hate']) > 0).astype(int)
train = train.drop(columns=['toxic', 'severe_toxic', 'severe_toxic', 'threat', 'insult', 'identity_hate', 'obscene', 'id'])
display(train[200:250]) 
```

```python
import contractions
import re
import unicodedata
import string

from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer() 

#remove http links
train['comment_text'] = train['comment_text'].apply(lambda text: re.sub(r'http\S+', '', text))
#convert accented characters
train['comment_text'] = train['comment_text'].apply(lambda text: unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore'))
#expand contractions
train['comment_text'] = train['comment_text'].apply(lambda text: contractions.fix(text))
#remove special characters
train['comment_text'] = train['comment_text'].apply(lambda text: re.sub(r'[^a-zA-z0-9.,!?/:;\"\'\s]', '', text))
#remove numbers
train['comment_text'] = train['comment_text'].apply(lambda text:  re.sub(r'[^a-zA-z.,!?/:;\"\'\s]' , '', text))
#remove punctation
train['comment_text'] = train['comment_text'].apply(lambda text: ''.join([(c if c not in string.punctuation else " ")for c in text]))
#remove extra white spaces
train['comment_text'] = train['comment_text'].apply(lambda text: re.sub(r'^\s*|\s\s*', ' ', text).strip())
#lower case
train['comment_text'] = train['comment_text'].apply(lambda text: ''.join(word.lower() for word in text))

display(train[200:250]) 
#https://towardsdatascience.com/nlp-building-text-cleanup-and-preprocessing-pipeline-eba4095245a0
```
